```
import torch
import torch.nn as nn

import torch.nn.functional as F


class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(
                in_channels=4,  # input height
                out_channels=32,  # n_filters
                kernel_size=5,  # filter size
                stride=1,  # filter movement/step
                padding=2,  # 如果想要 con2d 出来的图片长宽没有变化, padding=(kernel_size-1)/2 当 stride=1
            ),  # output shape (16, 28, 28)
            nn.ReLU(),  # activation
            nn.MaxPool1d(kernel_size=10),  # 在 2x2 空间里向下采样, output shape (16, 14, 14)
            # nn.BatchNorm1d(32)
        )
        self.conv2 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(32 ,64, 5, 1, 2),  # output shape (32, 14, 14)
            nn.ReLU(),  # activation
            nn.MaxPool1d(10),  # output shape (32, 7, 7)
            # nn.BatchNorm1d(64)

        )

        self.conv3 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(64 ,128, 5, 1, 2),  # output shape (32, 14, 14)
            nn.ReLU(),  # activation
            nn.MaxPool1d(10),  # output shape (32, 7, 7)
            # nn.BatchNorm1d(128)

        )
        self.out = nn.Linear(128 , 128)  # fully connected layer, output 10 classes
        self.out2 = nn.Linear(128 ,128)  # fully connected layer, output 10 classes
        self.out3 = nn.Linear(128, 1)  # fully connected layer, output 10 classes
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)  # 展平多维的卷积图成 (batch_size, 32 * 7 * 7)
        output = self.out(x)
        output = self.act(output)
        output = self.out2(output)
        output = self.act(output)
        output = self.out3(output)
        output = self.act(output)
        return output
```
