```
import torch
import torch.nn as nn

import torch.nn.functional as F


class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(
                in_channels=4,  # input height
                out_channels=64,  # n_filters 32
                kernel_size=5,  # filter size
                stride=1,  # filter movement/step
                padding=2,  # 如果想要 con2d 出来的图片长宽没有变化, padding=(kernel_size-1)/2 当 stride=1
            ),  # output shape (16, 28, 28)
            nn.ReLU(),  # activation
            nn.MaxPool1d(kernel_size=10),  # 在 2x2 空间里向下采样, output shape (16, 14, 14)
#             nn.BatchNorm1d(32)
        )
        self.conv2 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(64 ,128, 5, 1, 2),  # output shape (32, 14, 14) 32 64
            nn.ReLU(),  # activation
            nn.MaxPool1d(10),  # output shape (32, 7, 7)
#             nn.BatchNorm1d(64)

        )

        self.conv3 = nn.Sequential(  # input shape (1, 28, 28)
            nn.Conv1d(128 ,256, 5, 1, 2),  # output shape (32, 14, 14) 64 128
            nn.ReLU(),  # activation
            nn.MaxPool1d(10),  # output shape (32, 7, 7)
#             nn.BatchNorm1d(128)

        )
        self.out = nn.Linear(256 , 256)  # fully connected layer, output 10 classes
        self.out2 = nn.Linear(256 ,256)  # fully connected layer, output 10 classes
        self.out3 = nn.Linear(256, 1)  # fully connected layer, output 10 classes
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)  # 展平多维的卷积图成 (batch_size, 32 * 7 * 7)
#         x = 10*torch.log10(1/torch.sum(torch.abs(x,dim=1,keepdim=True))
        output = self.out(x)
        output = self.act(output)
        output = self.out2(output)
        output = self.act(output)
        output = self.out3(output)
        output = self.act(output)
        return output
```

```
import torch
from fastai.vision import cnn_learner, ItemList
# from scipy import misc
import joblib
import numpy as np
# import torchvision.transforms as transforms
from fastai.vision import cnn_learner, ItemList
# from torchvision.models import resnet18

# from try_cnn_snr.mynetwork import CNN
# from mynetwork import CNN
from fastai.train import  Learner

def get_df():
    all_information = joblib.load('fea').real.astype(np.float32)
    all_information2 = np.ones((all_information.shape[0],8,1025)).astype(np.float32)

    for batch_index,batch in enumerate(all_information):
        all_information2[batch_index,0:4,:] = all_information[batch_index]

        first =  np.fft.fft(all_information[batch_index,0,:-1],axis=-1)
        second = np.fft.fft(all_information[batch_index,1,:-1],axis=-1)
        all_information2[batch_index,4,:-1] = (first.real-first.real.mean())/first.real.std()
        all_information2[batch_index,5,:-1] =  (first.imag - first.imag.mean())/first.imag.std()
        all_information2[batch_index,6,:-1] = (second.real - second.real.mean())/second.real.std()
        all_information2[batch_index,7,:-1] = (second.imag - second.imag.mean())/second.imag.std()
        all_information2[batch_index,4:,-1] = all_information2[batch_index,0,-1]

    
    return all_information

if __name__ == '__main__':

    data = ItemList.from_array(get_df()).split_by_rand_pct()
    data = data.label_from_array().databunch()
    # print('xixi')
    # .label_from_func(lambda data:data[0,-1]).databunch()

    learner = Learner(data, CNN())
    # learner.model.double()
    # learner.lr_find()
    learner.fit(700,lr = 0.0001/4)
    learner.save('model3')
    print('xixi')
    # CNN
```
